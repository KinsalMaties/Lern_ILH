```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load data from CSV files into different Excel sheets
score_matrix = pd.read_csv('X_reduced_428.csv', header=None, delimiter=';')  # Score matrix
loading_matrix = pd.read_csv('X_loadings_428.csv', header=None, delimiter=';')  # Loading matrix
```

```python
# Step 2: Check if any values are strings and replace commas with dots
score_matrix = score_matrix.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)
loading_matrix = loading_matrix.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)

# Convert all elements to floats
score_matrix = score_matrix.astype(float)
loading_matrix = loading_matrix.astype(float)

# Step 3: Transpose the loading matrix
transposed_loading_matrix = loading_matrix.transpose()

# Step 4: Reconstruct the image using matrix multiplication
reconstructed_image = score_matrix.dot(transposed_loading_matrix)

# Step 5: Save the reconstructed image to a file
reconstructed_image.to_csv('reconstructed_image.csv', index=False, header=False)
```

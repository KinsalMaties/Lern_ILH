```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Шаг 1: Загрузка данных из CSV файлов в различные листы Excel
score_matrix = pd.read_csv('X_reduced_536.csv', header=None, delimiter=';')  # Матрица оценок
loading_matrix = pd.read_csv('X_loadings_536.csv', header=None, delimiter=';')  # Матрица загрузок
```

```python
# Шаг 2: Проверка наличия строковых значений и замена запятых на точки
score_matrix = score_matrix.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)
loading_matrix = loading_matrix.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)

# Преобразование всех элементов в числа с плавающей запятой
score_matrix = score_matrix.astype(float)
loading_matrix = loading_matrix.astype(float)
```

```python
# Шаг 3: Транспонирование матрицы загрузок
transposed_loading_matrix = loading_matrix.transpose()
```

```python
# Шаг 4: Восстановление изображения с помощью умножения матриц
reconstructed_image = score_matrix.dot(transposed_loading_matrix)
```

```python
# Шаг 5: Сохранение восстановленного изображения в файл
reconstructed_image.to_csv('reconstructed_image.csv', index=False, header=False)

reconstructed_image
```

```python
# Загрузка данных из CSV файла
reconstructed_image = pd.read_csv('reconstructed_image.csv', header=None)

# Преобразование данных в numpy array
image_array = reconstructed_image.to_numpy()

# Отображение изображения
plt.imshow(image_array, cmap='gray')
plt.axis('off')  # Отключение осей
plt.show()
```
